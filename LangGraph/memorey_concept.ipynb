```python
from langgraph.graph import StateGraph,START,END,MessagesState
from langchain_ollama import ChatOllama
from langgraph.checkpoint.memory import InMemorySaver
```

```python
model=ChatOllama(model="qwen2.5:7b",temperature=0.2)
```

```python
def call_model(state:MessagesState):
    response=model.invoke(state['messages'])
    return {"messages":[response]}
```

```python
checkpointer=InMemorySaver()
graph=StateGraph(MessagesState)

graph.add_node("call_model",call_model)
graph.add_edge(START,"call_model")
graph.add_edge('call_model',END)
```

```python
app=graph.compile(checkpointer=checkpointer)
```

```python
config={"configurable":{"thread_id":"thread-1"}}
config1={"configurable":{"thread_id":"thread-2"}}
```

```python
app.invoke({"messages":[{"role":"user","content":"Hi my age is 12."}]},config=config1)
```

```python
app.invoke({"messages":[{"role":"user","content":"hi what is my age"}]},config=config1)
```

```python
snap=app.get_state(config1)
vals=snap.values
for m in vals.get("messages",[]):
    print('-',type(m).__name__,":",m.content)
```
